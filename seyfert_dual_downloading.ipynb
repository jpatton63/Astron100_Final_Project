```python
import os
import pandas as pd
from astropy.coordinates import SkyCoord
import astropy.units as u
from astroquery.sdss import SDSS

excel_file = "seyfert 1 objects.xlsx"
df = pd.read_excel(excel_file)

df.columns = [col.strip() for col in df.columns]

output_dir = "sdss_spectra_seyfert"
os.makedirs(output_dir, exist_ok=True)

for idx, row in df.iterrows():
    name = row['Object'].strip()
    ra_str = str(row['RA']).strip()
    dec_str = str(row['DEC']).strip()

    print(f"ğŸ” Searching for {name} at RA: {ra_str}, DEC: {dec_str}")

    try:
        coord = SkyCoord(ra_str, dec_str, unit=(u.hourangle, u.deg), frame='icrs')

        xid = SDSS.query_region(coord, radius=2*u.arcsec, spectro=True)

        if xid is not None and 'plate' in xid.colnames:
            print(f"âœ… Found spectrum for {name}")
            spectra = SDSS.get_spectra(matches=xid)

            if spectra:
                fits_path = os.path.join(output_dir, f"{name.replace(' ', '_')}.fits")
                spectra[0].writeto(fits_path, overwrite=True)
                print(f"ğŸ’¾ Saved to {fits_path}")
            else:
                print(f"âš ï¸ Spectrum metadata found, but no FITS file available.")
        else:
            print(f"âŒ No spectroscopic data found for {name}")
    except Exception as e:
        print(f"ğŸš¨ Error for {name}: {e}")
```
