```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\jimmy\OneDrive\Astron100\BLL\ZBLLAC_1ES_0033_595_9_spectrum.txt" 

spectrum = pd.read_csv(file_path, 
                       delim_whitespace=True, 
                       comment='#', 
                       names=["Wavelength", "Flux"])

plt.figure(figsize=(10, 5))
plt.plot(spectrum["Wavelength"], spectrum["Flux"], color='navy')
plt.xlabel("Wavelength (Å)")
plt.ylabel("Flux (λ)")
plt.title("Spectral Plot")
plt.grid(True)
plt.tight_layout()
plt.show()
```
